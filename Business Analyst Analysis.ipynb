# Tranform some columns to extract new columns with difeerent information

1. Get minimum and maximum salary from Salary Estimate

```python
#df['Minimum Salary'] = 
df['Minimum Salary'] = df['Salary Estimate'].str[:4]
df['Minimum Salary'] = df['Minimum Salary'].str.replace(r'$', '', regex=True)#.astype('int')
df['Minimum Salary'] = df['Minimum Salary'].str.replace(r'[K]', '', regex=True)#.astype('int')
#df['Minimum Salary'] = re.sub('[^0-9.]', '', power_bank_price)
#df['Minimum Salary'] = df['Minimum Salary'].ljust(5,'0')
df['Minimum Salary'] = df['Minimum Salary'].str.pad(width=5, side='right', fillchar='0')
df['Minimum Salary'] = df['Minimum Salary'] .astype(int)
df['Minimum Salary'].value_counts()
```

```python
''' Maximum Salary ''' 
df['Maximum Salary'] = df['Salary Estimate'].str[6:10]
df['Maximum Salary'] = df['Maximum Salary'].str.replace(r'$', '', regex=True)
df['Maximum Salary'] = df['Maximum Salary'].str.replace(r'[K]', '', regex=True)
df['Maximum Salary'] = df['Maximum Salary'].str.replace(r' ', '', regex=True)
df['Maximum Salary'] = df['Maximum Salary'].str.replace(r'(', '', regex=True)
df['Maximum Salary']= df['Maximum Salary'].str.pad(width=6, side='right', fillchar='0')
df['Maximum Salary'] = df['Maximum Salary'].astype(int)
df['Maximum Salary'].head()
```

```python
df.isna().sum()
```

2. Split location of Office and Headquarters into City and State 

```python
''' Split the location into street and state, save into individual columns '''
df[['Office City', 'Office State']] = df['Location'].str.split(',', n=1, expand=True)
```

```python
''' Split the headquarters into street and state,  save into individual columns '''
df[['Headquarters City', 'Headquarters State']] = df['Headquarters'].str.split(',', n=1, expand=True)
df
```

Q6. Customer Satisfaction?

    1. Based on ratings

```python
''' Convert to NaN as float to be used in Tableau '''
df.loc[df['Rating'] == '-1.0', 'Rating'] = '-1'
df.loc[df['Rating'] == '-1', 'Rating'] = 'NaN'
```

```python
df['Rating'].value_counts()
```

```python
''' Convert the columns with numeric values to float and int instead of object '''
df['Rating'] = df['Rating'].astype(float)
df['Minimum Salary'] = df['Minimum Salary'].astype(int)
df['Maximum Salary'] = df['Maximum Salary'].astype(int)
```

```python
df.info()
```

```python
''' The datatype needs to be fixed before performing this '''
conditions = [
    (df['Rating'] <= 3),
    (df['Rating'] > 3) & (df['Rating'] <= 4),
    (df['Rating'] > 4) & (df['Rating'] <= 5)
]

values = ['Low rated', 'Medium rated', 'High rated']

df['Rating Evaluation'] = np.select(conditions, values)
df.head()
```

```python
''' Remove columns with too less and unnecessary information'''
df = df.drop('Job Description', axis=1)
df = df.drop('Competitors', axis=1)
df = df.drop('Easy Apply', axis=1)
df
```
